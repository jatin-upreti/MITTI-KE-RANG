```python
plt.figure(figsize=(15,10))
sns.heatmap(df1.isnull())                       #plot heatmap to visualize NaN
plt.show()
```

CALCULATED PERCENTAGE OF NULL VALUES

```python
percentage_null=df1.isnull().sum()/df1.shape[0]*100
percentage_null
```

LISTING COLUMNS HAVING NaN GREATER THAN 20%

```python
dr=percentage_null[percentage_null>20].keys()
dr
```

```python
#drop rows
df2=df1.drop(columns=dr)
df2
```

```python
#only 1 column left Nan with 18 Nan values
df2.isnull().sum()
```

```python
df2['self_employed'].value_counts()     #Checking the values in 'self_employed' column.
```

```python
#fill Nan values :) 

df2['self_employed'].fillna(df2['self_employed'].mode()[0],inplace=True)    
```

```python
df2.isnull().sum()
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(df2.isnull())
```
